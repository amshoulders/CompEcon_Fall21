```python
import sklearn
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.tree import export_text
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

import sklearn.feature_selection
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import LinearRegression
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
biden = pd.read_csv('biden.csv')
```

# Part 1

```python
# create x and y
y = biden['biden'].values
X = biden[['female', 'age', 'educ', 'dem', 'rep']].values
X_col = ['female', 'age', 'educ', 'dem', 'rep']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=25)
```

```python
tree = DecisionTreeRegressor(max_depth = 3, min_samples_leaf = 5)
fit_tree = tree.fit(X_train, y_train)
show = export_text(fit_tree, feature_names = X_col)
print(show)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)
tree.plot_tree(fit_tree, feature_names = X_col, filled = True)
```

```python
MSE = mean_squared_error(y_test, fit_tree.predict(X_test))
print('The SSE of tree of depth 3 and ', fit_tree.get_n_leaves(), ' leaves = ', MSE)
```

# Part 2

```python
from scipy.stats import randint as sp_randint
param_dist = {"max_depth": [3, 10],
              "min_samples_split": sp_randint(2, 20),
              "min_samples_leaf": sp_randint(2, 20)}
```

```python
optimal_param = sklearn.model_selection.RandomizedSearchCV(fit_tree, 
                                           param_dist, 
                                           n_iter=100, 
                                           scoring='neg_mean_squared_error', 
                                           n_jobs=-1, 
                                           cv=5, 
                                           random_state=25)
```

```python
optimal_param.fit(X,y).best_params_
```

```python
optimal_param.fit(X,y).best_score_
```

# Part 3

```python
tree_rf = RandomForestRegressor(
    max_depth=3,
    n_estimators=100,
    bootstrap=True,
    n_jobs=-1,
    oob_score=True,
    random_state=25)
tree_rf.fit(X, y)
```

```python
y_pred_rf = tree_rf.oob_prediction_
MSE_rf = mean_squared_error(y, y_pred_rf)
print('MSE=', MSE_rf)
```
